#### b. Correlation HeatMap

```python
# lightest and darkest cells are most correlated ones

plt.figure(figsize=(16, 12))
sns.heatmap(
    data=DataFrame.corr(),
    annot=True,
    cmap=sns.color_palette("magma"),
    linewidths=1,
    fmt=".2f",
    linecolor="white",
    annot_kws={"size": 16}
)

plt.title("Correlation HeatMap", fontsize=24)
plt.tight_layout()
plt.show()
```

#### c. Comparing Mean & Standard Deviation between attributes for target classes

```python
mean = DataFrame.groupby("target_class")[
    [' Mean of the integrated profile',
     ' Standard deviation of the integrated profile',
     ' Excess kurtosis of the integrated profile',
     ' Skewness of the integrated profile',
     ' Mean of the DM-SNR curve',
     ' Standard deviation of the DM-SNR curve',
     ' Excess kurtosis of the DM-SNR curve',
     ' Skewness of the DM-SNR curve']
].mean().reset_index()
mean = mean.drop("target_class", axis=1)

compare_mean = mean.transpose().reset_index()
compare_mean = compare_mean.rename(
    columns={
        'index': "features",
        0: "not_star",
        1: "star"
    })

compare_mean
```

```python
std = DataFrame.groupby("target_class")[
    [' Mean of the integrated profile',
     ' Standard deviation of the integrated profile',
     ' Excess kurtosis of the integrated profile',
     ' Skewness of the integrated profile',
     ' Mean of the DM-SNR curve',
     ' Standard deviation of the DM-SNR curve',
     ' Excess kurtosis of the DM-SNR curve',
     ' Skewness of the DM-SNR curve']
].std().reset_index()
std = std.drop("target_class", axis=1)

compare_std = std.transpose().reset_index()
compare_std = compare_std.rename(
    columns={
        'index': "features",
        0: "not_star",
        1: "star"
    })
compare_std
```

```python
mean.plot(kind="bar", width=.6, figsize=(13, 6), colormap="Set2")
plt.grid(True, alpha=.3)
plt.title("COMPARING MEAN OF ATTRIBUTES FOR TARGET CLASSES")

std.plot(kind="bar", width=.6, figsize=(13, 6), colormap="Set2")
plt.grid(True, alpha=.3)
plt.title("COMPARING STANDARD DEVIATION OF ATTRIBUTES FOR TARGET CLASSES")

plt.show()
```

```python
print ("[GREEN == STAR 1, RED == NOT STAR 0]")

plt.figure(figsize=(13, 14))

plt.subplot(2, 1, 1)
sns.pointplot(x="features", y="not_star", data=compare_mean, color="r")
sns.pointplot(x="features", y="star", data=compare_mean, color="g")

plt.xticks(rotation=60)
plt.xlabel("")
plt.ylabel("")
plt.grid(True, alpha=.3)
plt.title("COMPARING MEAN OF ATTRIBUTES FOR TARGET CLASSES")

plt.subplot(2, 1, 2)
sns.pointplot(x="features", y="not_star", data=compare_std, color="r")
sns.pointplot(x="features", y="star", data=compare_std, color="g")

plt.xticks(rotation=60)
plt.xlabel("")
plt.ylabel("")
plt.grid(True, alpha=.3)
plt.title("COMPARING STANDARD DEVIATION OF ATTRIBUTES FOR TARGET CLASSES")

plt.subplots_adjust(hspace=.4)
plt.show()
```

## B. Data PreProcessing

### 1. Splitting the Feature and Target fields

```python
targets = DataFrame.target_class.values
print("Labels: %s\n" % targets)

features = DataFrame.drop(
    ["target_class"],
    axis=1,
)
print("Features: \n%s" % features)
```

### 2. Scaling the Feature

```python
from sklearn.preprocessing import MinMaxScaler

minMaxScaler = MinMaxScaler(feature_range=(0, 1))
features_scaled_minmax = minMaxScaler.fit_transform(features)
print("Features Scaled By MinMax: \n%s \n" % features_scaled_minmax)
```

###  3. Splitting the Train and the Test rows

```python
from sklearn.model_selection import train_test_split

# train 75%, test 25%
x_train, x_test, y_train, y_test = train_test_split(
    features_scaled_minmax,
    targets,
    test_size=0.25
)
```

## C. Machine Learning Models



```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import GridSearchCV
```

### 1. Logistic Regression

> Hyper-Params to tune: 'C'

```python
from sklearn.linear_model import LogisticRegression

# Setup the hyperparameter grid
logreg_param = {
    'random_state': [42],
    'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
    'solver': ["liblinear"],
    'penalty': ["l1"]
}

logreg = LogisticRegression()
logreg_cv = GridSearchCV(logreg, logreg_param, cv=5)
logreg_cv.fit(x_train, y_train)

print("Best Logistic Regression Estimator: {}\n".format(logreg_cv.best_estimator_))
print("Best Logistic Regression Score: {}".format(logreg_cv.best_score_))
```

```python
# Use model from best estimator
lr_model = logreg_cv.best_estimator_

lr_model.fit(x_train, y_train)
y_head_lr = lr_model.predict(x_test)

lr_accuracy = accuracy_score(y_test, y_head_lr)
lr_report = classification_report(y_test, y_head_lr)
cm_lr = confusion_matrix(y_test, y_head_lr)

print("Logistic Regression Accuracy: {}\n".format(lr_accuracy))
print("Logistic Regression Report: \n{}".format(lr_report))
```

### 2. Decision Tree Classifier

> Hyper-Params to tune: 'max_depth'

```python
from sklearn.tree import DecisionTreeClassifier

# Setup the hyperparameter grid
dectree_param = {
    'random_state': [42],
    'max_depth': np.arange(1, 100)
}

dectree = DecisionTreeClassifier()
dectree_cv = GridSearchCV(dectree, dectree_param, cv=5)
dectree_cv.fit(x_train, y_train)

print("Best Decision Tree Classifier Estimator: {}\n".format(dectree_cv.best_estimator_))
print("Best Decision Tree Classifier Score: {}".format(dectree_cv.best_score_))
```

```python
# Use model from best estimator
dc_model = dectree_cv.best_estimator_

dc_model.fit(x_train, y_train)
y_head_dc = dc_model.predict(x_test)

dc_accuracy = accuracy_score(y_test, y_head_dc)
dc_report = classification_report(y_test, y_head_dc)
cm_dc = confusion_matrix(y_test, y_head_dc)

print("Decision Tree Classifier Accuracy: {}\n".format(dc_accuracy))
print("Decision Tree Classifier Report: \n{}".format(dc_report))
```

### 3. Random Forest Classifier

> Hyper-Params to tune: 'n_estimators'

```python
from sklearn.ensemble import RandomForestClassifier

# Setup the hyperparameter grid
ranfor_param = {
    'n_estimators': np.arange(1, 100),
    'random_state': [42],
    'max_leaf_nodes': [200],
    'criterion': ["gini", "entropy"]
}

ranfor = RandomForestClassifier()
ranfor_cv = GridSearchCV(ranfor, ranfor_param, cv=5)
ranfor_cv.fit(x_train, y_train)

print("Best Random Forest Classifier Estimator: {}\n".format(ranfor_cv.best_estimator_))
print("Best Random Forest Classifier Score: {}".format(ranfor_cv.best_score_))
```

```python
# Use model from best estimator
rfc_model = ranfor_cv.best_estimator_

rfc_model.fit(x_train, y_train)
y_head_rfc = rfc_model.predict(x_test)

rfc_accuracy = accuracy_score(y_test, y_head_rfc)
rfc_report = classification_report(y_test, y_head_rfc)
cm_rfc = confusion_matrix(y_test, y_head_rfc)

print("Random Forest Classifier Accuracy: {}\n".format(rfc_accuracy))
print("Random Forest Classifier Report: \n{}".format(rfc_report))
```

### 4. Naive Bayes Classifier

> No Hyper-Params to tune~

```python
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()

nb_model.fit(x_train, y_train)
y_head_nb = nb_model.predict(x_test)

nb_accuracy = accuracy_score(y_test, y_head_nb)
nb_report = classification_report(y_test, y_head_nb)
cm_nb = confusion_matrix(y_test, y_head_nb)

print("Naive Bayes Classifier Estimator: {}\n".format(nb_model))
print("Naive Bayes Classifier Accuracy: {}\n".format(nb_accuracy))
print("Naive Bayes Classifier Report: \n{}".format(nb_report))
```

### 5. K Nearest Neighbor

> Hyper-Params to tune: 'n_neighbors', 'weights'

```python
from sklearn.neighbors import KNeighborsClassifier

# Setup the hyperparameter grid
knn_param = {
    'n_neighbors': np.arange(1, 10),
    'weights': ["uniform", "distance"]
}

knn = KNeighborsClassifier()
knn_cv = GridSearchCV(knn, knn_param, cv=5)
knn_cv.fit(x_train, y_train)

print("Best K Nearest Neighbor Estimator: {}\n".format(knn_cv.best_estimator_))
print("Best K Nearest Neighbor Score: {}".format(knn_cv.best_score_))
```

```python
# Use model from best estimator
knn_model = knn_cv.best_estimator_

knn_model.fit(x_train, y_train)
y_head_knn = knn_model.predict(x_test)

knn_accuracy = accuracy_score(y_test, y_head_knn)
knn_report = classification_report(y_test, y_head_knn)
cm_knn = confusion_matrix(y_test, y_head_knn)

print("K Nearest Neighbor Accuracy: {}\n".format(knn_accuracy))
print("K Nearest Neighbor Report: \n{}".format(knn_report))
```

### 6. Support Vector Machine

> Hyper-Params to tune: 'C'

```python
from sklearn.svm import SVC

# Setup the hyperparameter grid
svm_param = {
    'random_state': [42],
    'C': np.arange(50, 300, 25),
    'gamma': [1.6],
    'kernel': ["poly"],
    'probability': [True]
}

svm = SVC()
svm_cv = GridSearchCV(svm, svm_param, cv=5)
svm_cv.fit(x_train, y_train)

print("Best Support Vector Machine Estimator: {}\n".format(svm_cv.best_estimator_))
print("Best Support Vector Machine Score: {}".format(svm_cv.best_score_))
```

```python
# Use model from best estimator
svm_model = svm_cv.best_estimator_

svm_model.fit(x_train, y_train)
y_head_svm = svm_model.predict(x_test)

svm_accuracy = accuracy_score(y_test, y_head_svm)
svm_report = classification_report(y_test, y_head_svm)
cm_svm = confusion_matrix(y_test, y_head_svm)

print("Support Vector Machine Accuracy: {}\n".format(svm_accuracy))
print("Support Vector Machine Report: \n{}".format(svm_report))
```
